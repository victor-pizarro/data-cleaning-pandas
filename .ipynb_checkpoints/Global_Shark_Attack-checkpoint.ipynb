### Hipotesis

```python
'''
Hipotesis nÂº1

Al realizar la limpieza y manipulaciÃ³n del dataset, perderemos aproximadamente el 40% de este.


dataset inicial: 25723 x 24 (registros x campos)
dataset final:   6302  x 7  (registros x campos)

Como se observa, perdemos una gran cantidad de registros.

'''
num_reg_inicial=25723
num_reg_final=6302

ratio=num_reg_final/num_reg_inicial

ratio_redondeado=round(ratio,3)

print(f'Tras la limpieza y manipulaciÃ³n de datos nos hemos quedado con el {ratio_redondeado*100}% del dataset inicial \n Por lo tanto, nuestra hipotesis es erronea, hemos perdido el {(1-(ratio_redondeado))*100}%')
```

```python
'''
Hipotesis nÂº2
Consideramos que debido a la gran extensiÃ³n de costa, los tres paÃ­ses con mayor nÃºmero de ataques serÃ¡n en orden:
1. Estados Unidos - USA
2. Australia
3. Brasil

Descartamos paÃ­ses como Canada y Rusia ya que mucha de la costa perteneciente a dichos paÃ­ses se encuentran en unas latitudes extremas,
donde se hace practicamente inexistente la presencia de tiburones.

'''
tabla_frecuencia_paises=pd.crosstab(shark_attack_2['Country'],'Ataques')
tabla_frecuencia_paises.sort_values('Ataques',ascending=False)
```

```python
# Ejemplo 1
tabla_frecuencia_paises.sort_values('Ataques',ascending=False).head(7).plot.bar()
plt.savefig("hipotesis_1_vertical.jpg")
```

```python
# Hipotesis 3:
'''
White shark
Tiger shark
Bull shark
Wobbegong shark
Blacktip shark
Blue shark
'''

import numpy as np
import matplotlib.pyplot as plt
grupos = ['White', 'Tiger', 'Bull', 'Wobbegong','Blacktip','Blue']
Letal = [44,25,15,0,0,4]
No_letal = [114,48,36,20,15,10]
 
indice = np.arange(len(grupos))
 
## Se crean las primeras barras
plt.bar(indice, Letal, label='Letal')
 
## Se crean las segundas barras y se apilan sobre las primeras
plt.bar(indice, No_letal, label='No letal',  bottom=Letal)
 
plt.xticks(indice, grupos)
plt.ylabel("Cantidad de ataques")
plt.xlabel("Especies")
plt.title('Letalidad por especie')
plt.legend()
 
plt.show()

plt.savefig("letalidad_por_especie.jpg",bbox_inches='tight')
```
